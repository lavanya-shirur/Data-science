```python
import pandas as pd
data=pd.read_csv("Elections.csv")
```

```python
data.shape
data.describe()
```

```python
def freq_table(data,columns):
    for col in columns:
        print("\n"+col)
        print(data[col].value_counts())
    
columns=['AGE','NO_PENDING_CRIMINAL_CASES','PARTY','STATE_LITERACY_RATE','RE_CONTEST','WINNER','TOTAL_VOTES','VOTING_PERCENTAGE']
print(freq_table(data,columns))
    
```

```python
import matplotlib.pyplot as plt
def bar_plot(data,cols):
    for col in cols:
        fig=plt.figure(figsize=(10,10))
        ax=fig.gca()
        counts=data[col].value_counts()
        counts.plot.bar(ax=ax,color='blue')
        ax.set_title("election data"+col)
        ax.set_xlabel(col)
        ax.set_ylabel("frequency")
        plt.show()    
cols=['AGE','SEX','EDUCATION','NO_PENDING_CRIMINAL_CASES','STATE_LITERACY_RATE','WINNER']
bar_plot(data,cols)
```
